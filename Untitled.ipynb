```python
import numpy as np
import matplotlib.pyplot as plt

# Define the initial weights and learning rate
W0 = 10
W1 = 0.2
W2 = -0.75
learning_rate = 0.05

# Define the training data
X = np.array([[1, 0, 0],  # Input data point 1
              [1, 1, 0],  # Input data point 2
              [1, 0, 1],  # Input data point 3
              [1, 1, 1]]) # Input data point 4

# Define the target labels (AND gate logic)
y = np.array([0, 0, 0, 1])

# Define the step activation function
def step_function(x):
    return np.where(x >= 0, 1, 0)

# Initialize variables for tracking errors and epochs
errors = []
max_epochs = 1000

# Training loop
for epoch in range(max_epochs):
    predictions = step_function(W0*X[:,0] + W1*X[:,1] + W2*X[:,2])
    error = y - predictions
    squared_error = np.sum(error**2)
    errors.append(squared_error)
    
    # Update weights
    W0 += learning_rate * np.dot(error, X[:,0])
    W1 += learning_rate * np.dot(error, X[:,1])
    W2 += learning_rate * np.dot(error, X[:,2])
    
    # Check for convergence
    if squared_error <= 0.002:
        print(f"Converged after {epoch + 1} epochs.")
        break

# Plot epochs vs. error
plt.plot(range(len(errors)), errors)
plt.xlabel('Epochs')
plt.ylabel('Sum-Square Error')
plt.title('Epochs vs. Error')
plt.show()

# Print the final weights
print("Final weights after training:")
print("W0 =", W0)
print("W1 =", W1)
print("W2 =", W2)
```
